```python
import numpy as np
import matplotlib.pyplot as plt

def g(x, a):
    return (x / a) * np.exp(-(x / a))

a_values = [1, 2, 3, 4]
x = np.linspace(0, 10, 100)

for a in a_values:
    y = g(x, a)
    plt.plot(x, y, label=f'a={a}')

plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('g(x) = (x / a) * e^(-(x / a))')
plt.legend()
plt.grid(False)
plt.show()
```
