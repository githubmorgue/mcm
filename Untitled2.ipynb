```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# 定义微分方程函数
def myODE(W, F, r):
    dWdF = r * W
    return dWdF

# 定义初始条件
W0_values = [0.1, 0.2, 0.3]
W1_values = [0.9, 0.8, 0.7]

# 定义 F 的范围
F = np.linspace(0, 1, 100)

# 绘制三个图像并输出 r 值
for i in range(len(W0_values)):
    W0 = W0_values[i]
    W1 = W1_values[i]

    # 求解微分方程并计算参数 r
    W = odeint(myODE, W0, F, args=(W1,))
    r = np.log(W1 / W0) / F[-1]

    # 输出 r 值
    print(f"在情况 {i+1} 下，r 值为：{r:.2f}")

    # 绘制结果
    plt.plot(F, W, label=f'W0={W0}, W1={W1}, r={r:.2f}')

plt.xlabel('F')
plt.ylabel('W')
plt.title('Results of solving differential equations')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


# 定义微分方程函数
def myODE(W, F, r):
    # 计算 dW/dF
    dWdF = r * W
    return dWdF


# 定义初始条件
W0 = 0.1
W1 = 0.9

# 定义 F 的范围
F = np.linspace(0, 1, 100)

# 求解微分方程
W = odeint(myODE, W0, F,args=(W1,))
r = np.log(W1 / W0) / F[-1]
print(r)

# 绘制结果
plt.plot(F, W)
plt.xlabel('F')
plt.ylabel('W')
plt.title('Result of differential equation')
plt.show()
```
